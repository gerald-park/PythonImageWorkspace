```python
import pandas as pd 
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] > 185
```

```python
filter = df['키'] >= 185
df[filter] # 키 185이상만 출력
```

```python
df[~filter] #반대 출력
```

```python
df.loc[ filter ,'수학'] #키 185 이상사람들의 수학 점수만 표시
```

```python
df.loc[ filter ,['수학','과학']] #키 185 이상사람들의 수학,과학 점수만 표시
```

# 다양한 조건
& 그리고 (and)

```python
cond1 = df['키'] >= 185
cond2 = df['학교'] == '북산고'
df.loc[cond1 & cond2]
```

### | 또는 (or)

```python
cond1 = (df['키'] < 170) | (df['키'] > 200) # 키가 170보다 작거나, 200보다 큰 학생 데이터
df.loc[cond1]
```
